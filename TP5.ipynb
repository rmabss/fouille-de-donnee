```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Chargement des données IRIS
iris = load_iris()
X = iris.data

# Sélection de deux caractéristiques pour la visualisation
feature1 = 0
feature2 = 1
X_selected = X[:, [feature1, feature2]]

# Entraînement du modèle K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_selected)
y_kmeans = kmeans.predict(X_selected)

# Affichage du graphique d'entraînement
plt.figure(figsize=(10, 6))

# Affichage des points de données IRIS
plt.scatter(X_selected[:, 0], X_selected[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Affichage des centres des clusters
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)

plt.title('Clusters K-Means')
plt.xlabel(iris.feature_names[feature1])
plt.ylabel(iris.feature_names[feature2])
plt.show()
```
